```python
# MATRIX PLOTS
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values='passengers')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap = 'coolwarm')
```

```python
sns.clustermap(fp, cmap='coolwarm')
```

```python
sns.clustermap(fp, cmap='magma', standard_scale = 1) # normalised data (gaussian dist)
```
